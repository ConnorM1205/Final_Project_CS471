```python
df = pd.read_csv('datasets/player_mvp_stats_adv.csv')
```

```python
# Filter the data for the 2021 season only.
df_2021 = df[df['Year'] == 2021]

# Remove duplicate player entries for the 2021 season.
df_unique_2021 = df_2021.drop_duplicates(subset='Player', keep='first')

# Select relevant columns (PTS, AST, TRB).
df_top_players = df_unique_2021[['Player', 'PTS', 'AST', 'TRB']]

# Drop any rows with missing values.
df_top_players = df_top_players.dropna()

# Normalize the columns.
df_top_players['PTS_norm'] = (df_top_players['PTS'] - df_top_players['PTS'].min()) / (df_top_players['PTS'].max() - df_top_players['PTS'].min())
df_top_players['AST_norm'] = (df_top_players['AST'] - df_top_players['AST'].min()) / (df_top_players['AST'].max() - df_top_players['AST'].min())
df_top_players['TRB_norm'] = (df_top_players['TRB'] - df_top_players['TRB'].min()) / (df_top_players['TRB'].max() - df_top_players['TRB'].min())

# Calculate the score.
df_top_players['Score'] = df_top_players['PTS_norm'] + df_top_players['AST_norm'] + df_top_players['TRB_norm']

# Select the top 10 unique players based on the score.
top_10_unique_players_2021 = df_top_players.nlargest(10, 'Score')[['Player', 'PTS', 'AST', 'TRB', 'Score']]

# Plot a bar chart for the top 10 unique players based on their score.
plt.figure(figsize=(12, 6))
plt.barh(top_10_unique_players_2021['Player'], top_10_unique_players_2021['Score'], color='lightcoral')
plt.title("Top 10 Unique Players by Combined Score (PTS, AST, TRB) in 2021")
plt.xlabel("Combined Score")
plt.ylabel("Player")
plt.gca().invert_yaxis()  # Place the highest score at the top.
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()

# Display the top 10 unique players with their PTS, AST, TRB, and score
print(top_10_unique_players_2021)
```

```python
# Filter the data for the 2021 season and select only the top 10 unique players.
df_2021 = df[df['Year'] == 2021]
df_unique_top_10_players = df_2021[df_2021['Player'].isin(top_10_unique_players_2021['Player'])]

# Group by team and calculate the win/loss percentage and include player names.
team_performance = df_unique_top_10_players.groupby('Tm').agg(
    win_loss_percentage=('W/L%', 'mean'),
    players=('Player', lambda x: ', '.join(x.unique()))  # Concatenate unique player names for each team.
).reset_index()

# Sort teams by win/loss percentage.
team_performance_sorted = team_performance.sort_values(by='win_loss_percentage', ascending=False)

# Plot a bar chart for team win/loss percentage with player names next to each bar.
plt.figure(figsize=(12, 6))
bars = plt.barh(team_performance_sorted['Tm'], team_performance_sorted['win_loss_percentage'], color='lightblue')

# Annotate player names next to the bars.
for i, bar in enumerate(bars):
    player_names = team_performance_sorted['players'].iloc[i]
    plt.text(bar.get_width() + 0.01, bar.get_y() + bar.get_height() / 2, player_names, va='center', ha='left', fontsize=10, color='black')

plt.title("Team Win/Loss Percentage of Top 10 Players in 2021")
plt.xlabel("Win/Loss Percentage")
plt.ylabel("Team")
plt.gca().invert_yaxis()  # Place the team with the highest win/loss percentage at the top.
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()

# Display the team performance table with associated players.
print(team_performance_sorted[['Tm', 'win_loss_percentage', 'players']])
```
