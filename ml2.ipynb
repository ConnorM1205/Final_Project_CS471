```python
def find_ap(combo:pd.DataFrame):
    actual=combo.sort_values("Share", ascending=False).head(5)
    predicted=combo.sort_values("predictions", ascending=False)
    ps=[]
    found=0
    seen=1
    for i,r in predicted.iterrows():
        if r['Player'] in actual['Player'].values:
            found+=1
            ps.append(found/seen)
        seen+=1
        
    return sum(ps)/len(ps)
```

```python
def add_ranks(combo:pd.DataFrame)->pd.DataFrame:
    combo=combo.sort_values('Share',ascending=False)
    combo['Rk']=list(range(1,combo.shape[0]+1))
    combo=combo.sort_values('predictions', ascending=False)
    combo['Predicted_Rk']=list(range(1,combo.shape[0]+1))
    combo['Diff']=combo["Rk"]-combo['Predicted_Rk']
    return combo
```

```python
ranking=add_ranks(all_preds[1])
ranking[ranking['Rk']<6].sort_values('Diff', ascending=False)
```

```python
def backtest(stats, model, year, predictors):
    aps=[]
    all_preds=[]
    for year in years[5:]:
        train=stats[stats["Year"]<year]
        test=stats[stats["Year"]==year]
        model.fit(train[predictors],train["Share"])
        preds=reg.predict(test[predictors])
        preds=pd.DataFrame(preds,columns=['predictions'],index=test.index)
        combo=pd.concat([test[["Player","Share"]],preds],axis=1)
        combo=add_ranks(combo)
        all_preds.append(combo)
        aps.append(find_ap(combo))
    return sum(aps)/len(aps),aps,pd.concat(all_preds)
```

```python
stats['MVP_Vote'] = (stats['Share'] > 0).astype(int)
from sklearn.neural_network import MLPClassifier
train = stats[stats["Year"] < 2018]
test = stats[stats["Year"] == 2018]
```

```python
from sklearn.preprocessing import StandardScaler

# Select only numeric predictors
numeric_predictors = train[predictors].select_dtypes(include=["number"])

# Initialize and fit the scaler
scaler = StandardScaler()
scaler.fit(numeric_predictors)

# Scale the numeric columns in train and test datasets
train[predictors] = scaler.transform(train[predictors].select_dtypes(include=["number"]))
test[predictors] = scaler.transform(test[predictors].select_dtypes(include=["number"]))
```

```python
clf = MLPClassifier(solver = 'lbfgs', alpha=1e-5, hidden_layer_sizes=(5,2), random_state=1)
clf.fit(train[predictors], train["MVP_Vote"])
```

```python
test['Prediction'] = clf.predict_proba(test[predictors])[:,1]
test['Prediction_Class'] = clf.predict(test[predictors])
```

```python
test.sort_values("Prediction", ascending=False).head(10)
test[['Player', 'Share', 'MVP_Vote', 'Prediction_Class', 'Prediction']].sort_values(
    'Prediction', ascending=False
).head(10)
```

```python
import matplotlib.pyplot as plt

plt.scatter(test['Share'], test['Prediction'], alpha=0.6)
plt.xlabel('Actual MVP Share')
plt.ylabel('Predicted Probability')
plt.title('Actual vs Predicted MVP Probabilities')
plt.show()
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score

# Calculate metrics
accuracy = accuracy_score(test["MVP_Vote"], test["Prediction_Class"])
precision = precision_score(test["MVP_Vote"], test["Prediction_Class"])
recall = recall_score(test["MVP_Vote"], test["Prediction_Class"])

print(f"Accuracy: {accuracy}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
```
