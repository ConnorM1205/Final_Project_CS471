```python
merged_df = pd.merge(
    stats, 
    players_adv, 
    on=['Player', 'Year'], 
    how='outer', 
    suffixes=('_players', '_players_adv'), 
    indicator=True
)


print("\nRows that are only in one dataframe:")
print(merged_df[merged_df['_merge'] != 'both'])


merged_df = merged_df.drop(columns=['_merge'])


columns_players = [col for col in merged_df.columns if col.endswith('_players')]
columns_players_adv = [col.replace('_players', '_players_adv') for col in columns_players]


def find_differences(row):
    diffs = {}
    for col1, col2 in zip(columns_players, columns_players_adv):
        if row[col1] != row[col2]:
            diffs[col1] = row[col1]
            diffs[col2] = row[col2]
    return pd.Series(diffs)


differences = merged_df.apply(find_differences, axis=1).dropna(how='all')

print(f"\nRows with differences found: {len(differences)} rows.")
print(differences)
```
