```python
def rule_based(player: dict) -> int:
    mvp_score = 0

    # Rule 2: Scoring
    if player['PTS'] >= 30:
        mvp_score += 3
    elif player['PTS'] >= 25:
        mvp_score += 2
    elif player['PTS'] >= 20:
        mvp_score += 1

    # Rule 3: Rebounds
    if player['TRB'] >= 10:
        mvp_score += 2
    elif player['TRB'] >= 7:
        mvp_score += 1

    # Rule 4: Assists
    if player['AST'] >= 8:
        mvp_score += 2
    elif player['AST'] >= 5:
        mvp_score += 1
        
    # Rule 5: Triple Double
    if player['PTS']>=10 and player['AST']>=10 and player['TRB']>=10:
        mvp_score +=3

    # Rule 6: Stocks (Steals and Blocks combined)
    stocks = player['STL'] + player['BLK']
    if stocks >= 3:
        mvp_score += 2
    elif stocks >= 2:
        mvp_score += 1

    # Rule 7: Field Goal Percentage
    if player['FG%'] >= 0.51:
        mvp_score += 2
    elif player['FG%'] >= 0.45:
        mvp_score += 1

    if player['3PA']>=3 and player['3P%']>.39:
        mvp_score+=2
    elif player['3PA']>=3 and player['3P%']>.375:
        mvp_score+=1
        
    # Rule 8: Wins
    if player['W'] >= 50:
        mvp_score += 3
    elif player['W'] >= 45:
        mvp_score += 2
    elif player['W'] >= 40:
        mvp_score += 1

    # Rule 9: PER
    if player['PER'] >= 25:
        mvp_score += 2
    elif player['PER'] >= 22:
        mvp_score += 1
    
    # Rule 10: Win Shares per 48 minutes (WS/48)
    if player['WS/48'] >= 0.250:
        mvp_score += 3
    elif player['WS/48'] >= 0.200:
        mvp_score += 2
    elif player['WS/48'] >= 0.150:
        mvp_score += 1
    
    # Rule 11: Voter Fatigue
    if player['P_MVP']:
        mvp_score-=1
    return mvp_score
```
