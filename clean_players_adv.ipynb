```python
players['Player'] = players['Player'].str.replace('.', '', regex=False)
players_adv['Player'] = players_adv['Player'].str.replace('.', '', regex=False)
players_adv['Player'] = players_adv['Player'].replace({
    'Marko Gudurić': 'Marko Guduric',
    'KJ Martin': 'Kenyon Martin Jr'
})
merged_df = pd.merge(
    players, 
    players_adv, 
    on=['Player', 'Year'], 
    how='outer', 
    suffixes=('_players', '_players_adv'), 
    indicator=True
)

# Display rows that didn't have a match in either dataframe
print("\nRows that are only in one dataframe:")
print(merged_df[merged_df['_merge'] != 'both'])

# Drop the indicator column since it's not needed for comparison
merged_df = merged_df.drop(columns=['_merge'])

# Get lists of columns from each dataframe based on the suffixes
columns_players = [col for col in merged_df.columns if col.endswith('_players')]
columns_players_adv = [col.replace('_players', '_players_adv') for col in columns_players]

# Create a DataFrame to track differences between the two dataframes
def find_differences(row):
    diffs = {}
    for col1, col2 in zip(columns_players, columns_players_adv):
        if row[col1] != row[col2]:
            diffs[col1] = row[col1]
            diffs[col2] = row[col2]
    return pd.Series(diffs)

# Apply the comparison and extract rows with differences
differences = merged_df.apply(find_differences, axis=1).dropna(how='all')

print(f"\nRows with differences found: {len(differences)} rows.")
print(differences)
```
