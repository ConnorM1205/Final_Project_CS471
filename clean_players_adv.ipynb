```python
del players_adv['Unnamed: 0']
del players_adv['Unnamed: 19']
del players_adv['Unnamed: 24']
del players_adv['Rk']
```

```python
players_adv['Player']=players_adv['Player'].str.replace('*','', regex=False)
```

```python
def combine_players(df):
    if df.shape[0] == 1:
        return df
    else:
        row = df[df["Tm"] == "TOT"].copy()
        row["Tm"] = df.iloc[-1,:]["Tm"]
        return row

players_adv = players_adv.groupby(["Player", "Year"]).apply(combine_players)
```

```python
players_adv.index=players_adv.index.droplevel()
```

```python
players_adv.index=players_adv.index.droplevel()
```
